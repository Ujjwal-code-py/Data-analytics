## Processing batting unit


```python
## 
```

```python
with open(r't20_json_files\t20_json_files\t20_wc_batting_summary.json') as d:
    data=json.load(d)
data
all_records=[]
for rec in data:
    all_records.extend(rec['battingSummary'])
all_records   
df_batting=pd.DataFrame(all_records) 
```

```python
match_id_dict={}
for index, row in df_match.iterrows():
    key1= row['team1']+ ' Vs '+row['team2']
    key2= row['team2']+ ' Vs '+row['team1']
    
    match_id_dict[key1] = row['matchid']
    match_id_dict[key2] = row['matchid']
    
match_id_dict
```

```python
df_batting.head(20)
```

```python
df_batting['out/not_out']=df_batting.dismissal.apply(lambda x: "out" if(len(x)>0) else "not_out")
```

```python
df_batting.head(20)
df_batting.drop(columns=['dismissal'],inplace=True)
df_batting.head(20)
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace("â€",""))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace("/xa0",""))
```

```python
df_batting.sample(30)
```

```python
match_id_dict['Pakistan Vs England']
```

```python
df_batting['matchid']=df_batting['match'].map(match_id_dict)
```

```python
df_batting.sample(5)
```

```python
df_batting.to_csv('match/batting.csv',index=False)
```
